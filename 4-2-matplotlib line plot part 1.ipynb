```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy.random import randint

x = randint(0,20,10)
y = np.linspace(0,50,10)
z = randint(0,30,10)

x,y,plt.plot(x,y) # blue line
x,y,plt.plot(y,x) # orange line
x,y,z,plt.plot(y,z) # green line
```
