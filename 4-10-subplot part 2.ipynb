###
##  Subplot Part1 Code
###

```python

import matplotlib.pyplot as plt

a = [1,2,3,4]
b = [9,2,5,3]
c = [10,19,7,18]

# subplot creates different plots in one cell without mixing them
plt.subplot(4,1,1)
plt.plot(a,b)

plt.subplot(4,1,2)
plt.plot(b,a)

plt.subplot(4,1,3)
plt.plot(a,c)

plt.subplot(4,1,4)
plt.plot(b,c)
```

###
## Subplot Part 2 
2x2 subplot

###

```python

import matplotlib.pyplot as plt

a = [1,2,3,4]
b = [9,2,5,3]
c = [19,12,18,11]

plt.subplot(2,2,1)
plt.plot(a,b,color = 'b', linestyle = '-', linewidth = 1, marker = 'o', markersize = 5, markerfacecolor = 'g', markeredgecolor = 'r', markeredgewidth = 3)
plt.xlabel('a (1,2,3,4)')
plt.ylabel('b (9,2,5,3)')
plt.title('\'a & b\'')

plt.subplot(2,2,2)
plt.plot(b,c,color = 'g', linestyle = '--', linewidth = 2, marker = 'v', markersize = 6, markerfacecolor = 'r', markeredgecolor = 'k', markeredgewidth = 4)
plt.xlabel('b (9,2,5,3)')
plt.ylabel('c (19,12,18,11)')
plt.title('\'b & c\'')

plt.subplot(2,2,3)
plt.plot(c,b, color = 'y', linestyle = ':', linewidth = 1.5, marker = 'h', markersize = 5.5, markerfacecolor = 'y', markeredgecolor = 'g', markeredgewidth = 5)
plt.xlabel('c (19,12,18,11)')
plt.ylabel('b (9,2,5,3)')
plt.title('\'c & b\'')

plt.subplot(2,2,4)
plt.plot(b,a, color = 'c', linestyle = '-', linewidth = 1.7, marker = 's', markersize = 6.1, markerfacecolor = 'b', markeredgecolor = 'y', markeredgewidth = 5.5)
plt.xlabel('b (9,2,5,3)')
plt.ylabel('a (1,2,3,4)')
plt.title('\'b & a\'')

plt.tight_layout() # it prevents overlaping
```
