###
## Subplots
In subplots we can define no. of rows & columns only once

###

```python

import matplotlib.pyplot as plt

a = [1,2,3,4]
b = [9,2,5,3]
c = [10,19,7,18]

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,6)) # in subplots we can define no. of rows & columns only once

ax[0,0].plot(a,b)

ax[0,1].plot(b,a)

ax[1,0].plot(b,c)

ax[1,1].plot(b,c)
```

###
## Subplots 
Setting labels

###

```python

import matplotlib.pyplot as plt

a = [1,2,3,4]
b = [9,2,5,3]
c = [10,19,7,18]

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,6)) # in subplots we can define no. of rows & columns only once

ax[0,0].plot(a,b)
ax[0,0].set_xlabel('a')
ax[0,0].set_ylabel('b')
ax[0,0].set_title('a & b')

ax[0,1].plot(b,a)

ax[1,0].plot(b,c)

ax[1,1].plot(c,b)

plt.tight_layout()
```
