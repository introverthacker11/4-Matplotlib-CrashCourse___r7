###
## Scatter Plot
###

```python

import matplotlib.pyplot as plt
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv')
b = a['College']
c = a['Rank']
plt.figure(figsize = (10,3))
a,plt.scatter(b,c)
```

###
## Bar Plot
###

```python

import matplotlib.pyplot as plt
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv')
b = a['College']
c = a['Rank']
plt.figure(figsize = (10,3))
a,plt.scatter(b,c)
a,plt.bar(b,c)
```

## Bar Plot Before Scatter Plot

```python

import matplotlib.pyplot as plt
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv')
b = a['College']
c = a['Age']
plt.figure(figsize = (10,3))
a,plt.bar(b,c)
a,plt.scatter(b,c)
```
