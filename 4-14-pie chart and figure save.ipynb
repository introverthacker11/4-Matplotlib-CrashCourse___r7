```python
# introverts food consumption plot

import matplotlib.pyplot as plt

a = ['Rayyan','Wajahat','Hamza','Moiz','Sami','Muzammil','Jumbo']
b = [35,16,20,12,10,18,82]

plt.plot(a,b, color = 'b', linestyle = '--', linewidth = 1, marker = 'o', markersize = 10, markerfacecolor = 'g', markeredgecolor = 'r', markeredgewidth = 3, label = 'Consumption rate in kg')
plt.xlabel('Introverts')
plt.ylabel('Food Consumed per week (in KG)')
plt.title('\"Plot Title: Introvert\'s\" vs \"Food Consumption\"\n')
plt.grid()
plt.legend()
```

###
## Saving pie chart 
###

```python

import matplotlib.pyplot as plt
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv').head(5)

plt.figure(figsize = (10,3))

print(a)
plt.pie(a['Rank'], labels = a['College'], autopct = '%1.1f%%', shadow = True, startangle = 90, explode = (0.3,0.4,0,0,0)) 
plt.savefig('piechart.png', dpi = 300)
plt.show()
```
