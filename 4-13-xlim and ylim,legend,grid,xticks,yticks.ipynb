###
## x_lim and y_lim
###

```python

import matplotlib.pyplot as plt

a = [10,30,20,50,45]
b = [15,35,40,30,10]
c = [12,19,27,32,45]

fig, ax =  plt.subplots(nrows = 2, ncols = 2, figsize = (10,3))

ax[0,0].plot(a,b)
ax[0,1].plot(b,a)
ax[1,1].plot(b,c)
# creating zoomed figure

fig = plt.figure(figsize = (10,3))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.1,0.3,0.5])

ax1.plot(b,a)
ax2.plot(b,a)

ax2.set_xlim(10,37)
ax2.set_ylim(15,46)
```

###
## legend  ( numpy )
###

```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.figure(figsize = (10,3))

rating = [11,35,14,20,15,19,39,23]
ratings = [r + 3 for r in rating]

plt.plot(rating, marker = 'o', label = 'ratings')
plt.plot(ratings, marker = 'o', label = 'ratings + 3')

plt.legend()
```

###
## xticks & yticks 
###

```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.figure(figsize = (10,3))

rating = [11,35,14,20,15,19,39,23]
ratings = [r + 3 for r in rating]

plt.plot(rating, marker = 'o', label = 'ratings')
plt.plot(ratings, marker = 'o', label = 'ratings + 3')

plt.xticks(np.arange(0,10,7))
plt.grid()
plt.legend()
```
