## Subplot
Subplot creates different plots in one cell without mixing them

###

```python

import matplotlib.pyplot as plt

a = [1,2,3,4]
b = [9,2,5,3]
c = [10,19,7,18]

plt.subplot(2,2,1)
plt.plot(a,b)

plt.subplot(2,2,2)
plt.plot(b,a)

plt.subplot(2,2,3)
plt.plot(a,c)
```

```python
# subplot

import matplotlib.pyplot as plt

a = [1,2,3,4]
b = [9,2,5,3]
c = [10,19,7,18]

# subplot creates different plots in one cell without mixing them
plt.subplot(1,4,1)
plt.plot(a,b)

plt.subplot(1,4,2)
plt.plot(b,a)

plt.subplot(1,4,3)
plt.plot(a,c)

plt.subplot(1,4,4)
plt.plot(c,b)
```
