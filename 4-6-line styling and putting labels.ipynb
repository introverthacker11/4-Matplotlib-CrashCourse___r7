###
## 1) Coloring plots  
There are different colours i.e: b,r,g,k,y,w,m,c

###

```python

import matplotlib.pyplot as plt
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv')
b = a['College']
c = a['Age']
plt.figure(figsize = (10,3))
a,plt.plot(b,c, color = 'c') 
```

###
## Labels
###

```python

import matplotlib.pyplot as plt

a = pd.read_csv('groupby(csvfile8).csv')
b = a['College']
c = a['Age']
plt.figure(figsize = (10,3))
a,plt.plot(b,c, color = 'b', linestyle = None, linewidth = 1, marker = 'o', markersize = 5, markerfacecolor = 'g', markeredgecolor = 'r', markeredgewidth = 3) 
plt.xlabel('\n\nCollege')
plt.ylabel('Age\n\n')
plt.title('College vs Age\n\n')
#  For instance, (0.0, 0.0) would be bottom-left, (1.0, 0.0) would be bottom-right, and (0.5, 0.5) 
# would be centered horizontally and vertically.
```
