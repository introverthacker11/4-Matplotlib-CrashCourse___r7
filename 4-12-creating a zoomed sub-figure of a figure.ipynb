###
## Subplots
###

```python

import matplotlib.pyplot as plt

a = [1,2,3,4]
b = [9,2,5,3]
c = [10,19,7,18]

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,6)) # in subplots we can define no. of rows & columns only once

ax[0,0].plot(a,b)

ax[0,1].plot(b,a)

ax[1,0].plot(b,c)

ax[1,1].plot(c,b)

```

###
## Zoomed Figure
###

```python

import matplotlib.pyplot as plt

a = [1,2,3,4]
b = [9,2,5,3]
c = [10,19,7,18]

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,6)) # in subplots we can define no. of rows & columns only once
ax[0,0].plot(a,b)
ax[1,1].plot(a,b)

# creating zoomed sub-figure
fig = plt.figure(figsize = (10,3))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.1,0.2,0.4])

ax1.plot(a,b)
ax2.plot(b,c)
```

###
## x and y limits
###

```python

import matplotlib.pyplot as plt

a = [1,2,3,4]
b = [9,2,5,3]
c = [10,19,7,18]

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,6)) # in subplots we can define no. of rows & columns only once
ax[0,0].plot(a,b)
ax[0,1].plot(b,c)
ax[1,0].plot(a,c)


# creating zoomed sub-figure
fig = plt.figure(figsize = (10,3))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.1,0.2,0.4])

ax1.plot(a,b)
ax2.plot(b,c)

ax2.set_xlim(2,5.5)
ax2.set_ylim(8,16)
```

```python
# x and y limits

import matplotlib.pyplot as plt

a = [1,7,19,23,30]
b = [10,11,18,20,1]
c = [11,17,48,8,3]

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,6))

ax[1,0].plot(a,b)
ax[0,0].plot(a,c)
ax[1,1].plot(b,a, marker = 'o')

# creating zoomed figure

fig = plt.figure(figsize = (10,3))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.1,0.2,0.4])

ax1.plot(b,a, marker = 'o')
ax2.plot(b,a, marker = 'o')

ax2.set_xlim(9,19)
ax2.set_ylim(0,26)
```
